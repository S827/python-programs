```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv('../../sales_subset.csv', index_col=0)
sales.head()
```

## 1•  Group By .pivot_table()

•	How would you create a pivot table to calculate the average weekly sales for each department?

•	Create a pivot table that shows the mean weekly sales for each store.

```python
sales.pivot_table(values='weekly_sales', index='department')
```

## 2•  Different Statistics

•	Create a pivot table to calculate the median weekly sales for each department.



```python
sales.pivot_table(values='weekly_sales', index='department', aggfunc=np.median)
```

•	How can you modify the pivot table to calculate the median weekly sales for each store?

```python
sales.pivot_table(values='weekly_sales', index='store', aggfunc=np.median)
```

## 3•  Multiple Statistics

•	Generate a pivot table that calculates both the mean and median weekly sales for each department.

•	Create a pivot table showing both the mean and sum of weekly sales for each store.

```python
# • Generate a pivot table that calculates both the mean and median weekly sales for each department.
sales.pivot_table(values='weekly_sales', index='department', aggfunc=[np.mean, np.median])
```

```python
# • Create a pivot table showing both the mean and sum of weekly sales for each store
sales.pivot_table(values='weekly_sales', index='store', aggfunc=[np.mean, np.sum])
```

## 4•  Pivot on Two Variables

•	How can you create a pivot table to calculate the average weekly sales for each combination of department and holiday status?

•	Create a pivot table to show the median weekly sales for each combination of store and holiday status.

```python
# • How can you create a pivot table to calculate the average weekly sales for each combination of department and holiday status?
sales.pivot_table(values='weekly_sales', index='department', columns='is_holiday')
```
