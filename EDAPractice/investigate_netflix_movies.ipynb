### 1. What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration.

Here, we need to look for 
1. movie, 
2. year and 
3. duration

```python
# filter by movie type
netflix_movies = netflix_df[netflix_df['type'] == 'Movie']
netflix_movies.head()
```

```python
#  filter by release_year from 1990 to 2000
netflix_movies_1990s = netflix_movies[np.logical_and(netflix_movies['release_year'] >= 1990, netflix_movies['release_year'] < 2000)]
netflix_movies_1990s.head()
```

```python
netflix_movies_1990s.info()
```

```python
netflix_movies_after_1990 = netflix_movies[netflix_movies['release_year'] >= 1990]
netflix_movies_before_2000 = netflix_movies_after_1990[netflix_movies_after_1990['release_year'] < 2000]
netflix_movies_before_2000.info()
```

```python
netflix_movies_1990s['release_year'] == netflix_movies_before_2000['release_year']
```

Now it is clear that filterd netflix_movies_1990s and netflix_movies_before_2000 are same but first one is more compact as code is one-liner.

Visualize the duration column of above filtered data

```python
plt.hist(netflix_movies_1990s['duration'])
plt.title('Distribution of Movie durations in 1990s')
plt.xlabel('Duration in Minutes')
plt.ylabel('Number of Movies')
plt.show()
```

With above histogram, we can say most frequent duration is 100 minutes

## 2. A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

1. need to count
2. action movies
3. short action movies(< 90 minutes)-- duration
4. released in 1990s

```python
# we already have 1990s movies dataframe, filter action movies from it
netflix_action_mov_1990s = netflix_movies_1990s[netflix_movies_1990s['genre'] == 'Action']
netflix_action_mov_1990s.head()
```

```python
# now we need to filter the movies which are less than 90 minutes
# till now what we have learnt is we can iterate over rows and count it
count = 0
for lab, row in netflix_action_mov_1990s.iterrows():
    if row['duration'] < 90:
        count += 1
    else:
        count = count
print(count)
```
