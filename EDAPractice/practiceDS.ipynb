### Problem Statement: Plot the daily temperatures of a city over a month using a line plot.

# Dataset 4: Quarterly Revenue
### Problem Statement: Plot the quarterly revenue of a company for the last 3 years.

```python
year = 2021
year_list = []
for y in range(10):
    for i in range(4):
        year_list.append("Q"+str(i+1)+ " " + str(year))
    year += 1
year_list_np = np.array(year_list)
year_list_np
revenue = np.round(np.random.normal(2, 1, 40), 1)
revenue
plt.xlabel("Quarters")
plt.ylabel("Revenue")
plt.grid(True)
plt.title("Quarterly Revenues")
plt.xticks(rotation=90)
plt.plot(year_list_np, revenue, marker="*");
```

we can see from above plot, revenue was maximum in 2021 4th quarters and 3rd quarter, and least in 2023 4th quarter and 3rd quarter

## Dataset 1: House Prices and Sizes
### Problem Statement: Create a scatter plot to visualize the relationship between house prices and their sizes.

```python
house_size = np.round(np.random.normal(1000, 300, 100), 0)
house_price = np.round(np.random.normal(3000000, 1000000, 100), 0)
print(house_size)
print(house_price)
```

```python
def plot(x_t, y_t, title, x, y):
    plt.xlabel(x_t)
    plt.ylabel(y_t)
    plt.title(title)
    plt.grid(True)
    plt.scatter(x, y)
```

```python
plot("house size in sq ft.", "house Price in millions", "House Prices and Sizes", house_size, house_price)
```

we can see clearly that there is no strong relationship between house price and house size as house with 600 sq ft costs more than 5 millions and house with 1750 sq ft costs 4 millions, there could be another factor which cause house prices up and low
