# 1. Problem: Daily temperatures
### Problem Statement: Plot the daily temperatures of a city over a month using a line plot.

```python
days = list(range(1, 32))
days_np = np.array(days)
temp = np.round(np.random.normal(30, 10, 31), 2)
temp.shape
days_np.shape
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title("Daily temperature over a month")
plt.grid(True)
plt.plot(days_np, temp, marker='x');
```
