```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import calendar
```

# 1. Problem: Daily temperatures
### Problem Statement: Plot the daily temperatures of a city over a month using a line plot.

```python
days = list(range(1, 32))
days_np = np.array(days)
temp = np.round(np.random.normal(30, 10, 31), 2)
temp.shape
days_np.shape
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title("Daily temperature over a month")
plt.grid(True)
plt.plot(days_np, temp, marker='x');
```

# Dataset 2: Monthly Rainfall
### Problem Statement: Use a line plot to show the monthly rainfall in a region for a year.

```python
months = np.array(list(calendar.month_abbr)[1:])
months.shape
rainfall = np.round(np.random.normal(90, 10, 12), 2)
rainfall
plt.xlabel("Month")
plt.ylabel("Rainfall")
plt.title("Monthly Rainfall")
plt.grid(True)
plt.plot(months, rainfall, marker='+');
```

# Dataset 3: Weekly Workout Hours
### Problem Statement: Plot the weekly workout hours of an individual over a period of 10 weeks.

```python
weeks = np.array(list(range(1,11)))
workout_hours = np.round(np.random.normal(5, 2, 10), 0)
workout_hours
plt.xlabel("Week Number")
plt.ylabel("Workout in Hours")
plt.title("Weekly Workout Hours")
plt.grid(True)
plt.plot(weeks, workout_hours, marker="*");
```

# Dataset 4: Quarterly Revenue
### Problem Statement: Plot the quarterly revenue of a company for the last 3 years.

```python
year = 2021
year_list = []
for y in range(3):
    for i in range(4):
        year_list.append("Q"+str(i+1)+ " " + str(year))
    year += 1
year_list_np = np.array(year_list)
year_list_np
revenue = np.round(np.random.normal(2, 1, 12), 1)
revenue
plt.xlabel("Quarters")
plt.ylabel("Revenue")
plt.grid(True)
plt.title("Quarterly Revenues")
plt.xticks(rotation=90)
plt.plot(year_list_np, revenue, marker="*");
```

we can see from above plot, revenue was maximum in 2021 4th quarters and 3rd quarter, and least in 2023 4th quarter and 3rd quarter
